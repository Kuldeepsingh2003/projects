# Column description
##ID: Customer’s unique identifier
##Year_Birth: Customer’s birth year
##Education: Customer’s education level
##Marital_Status: Customer’s marital status
##Kidhome: Number of children in customer’s household
##Tennhome: Number of teenagers in customer’s household
##Dt_Customer: Date of customer’s enrollment with the company
##Recency: Number of days since customer’s last purchase
##MntWines: Amount spent on wine in the last 2 years
##MntFruits: Amount spent on fruits in the last 2 years
##MntMeatProducts: Amount spent on meat in the last 2 years
##MntFishProducts: Amount spent on fish in the last 2 years
##MntSweetProducts: Amount spent on sweets in the last 2 years
##MntGoldProds: Amount spent on gold in the last 2 years
##NumDealsPurchase: Number of purchases made with a discount
##NumWebPurchase: Number of purchases made through the company’s website
##NumCatalogPurchase: Number of purchases made using a catalog
##NumStorePurchase: Number of purchases made directly in stores
##NumWebVisitsMonth: Number of visits to company’s website in the last month
##AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise
##AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise
##AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise
##AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 otherwise
##AcceptedCmp2: 2 if customer accepted the offer in the 1st campaign, 0 otherwise
##Response: 1 if customer accepted the offer in the last campaign, 0 otherwise
##Complain: 1 if a customer complained in the last 2 years, 0 otherwise
##Country: Customer’s location

```python
# Create a scatter plot to visualize the relationship between income and number of store purchases
plt.figure(figsize=(10, 6))
plt.scatter(df['Income'], df['NumStorePurchases'], alpha=0.5, edgecolors='k')
plt.title('Income vs. Number of Store Purchases')
plt.xlabel('Income')
plt.ylabel('Number of Store Purchases')
plt.grid(True)
plt.show()

# Create a box plot to show the distribution of store purchases across different income categories
income_bins = [0, 30000, 60000, 90000, 120000, 150000, 200000]
income_labels = ['0-30k', '30k-60k', '60k-90k', '90k-120k', '120k-150k', '150k+']
df['IncomeCategory'] = pd.cut(df['Income'], bins=income_bins, labels=income_labels, include_lowest=True)

plt.figure(figsize=(12, 6))
df.boxplot(column='NumStorePurchases', by='IncomeCategory', grid=False)
plt.title('Number of Store Purchases by Income Category')
plt.suptitle('')  # Suppress the default title to keep only the custom one
plt.xlabel('Income Category')
plt.ylabel('Number of Store Purchases')
plt.show()
```

```python


# Define age bins and labels
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, 100]
age_labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']

# Categorize ages into bins
df['AgeGroup'] = pd.cut(df['Customer Age'], bins=age_bins, labels=age_labels, right=False)

# Group by AgeGroup and sum the NumStorePurchases
age_store_purchases = df.groupby('AgeGroup')['NumStorePurchases'].sum()

# Plot the distribution of store purchases by age group
plt.figure(figsize=(10, 6))
age_store_purchases.plot(kind='bar', color='skyblue', edgecolor='k')
plt.title('Distribution of Store Purchases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Store Purchases')

plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the columns of interest
columns = ['NumDealsPurchases', 'NumWebPurchases',
           'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

# Create subplots
fig, axes = plt.subplots(1, len(columns), figsize=(20, 5))

# Plot histograms for each column
for i, column in enumerate(columns):
    sns.histplot(df[column], ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the columns of interest
columns = ['NumDealsPurchases', 'NumWebPurchases',
           'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

# Create subplots
fig, axes = plt.subplots(1, len(columns), figsize=(20, 7))

# Plot boxplots for each column, grouped by income category
for i, column in enumerate(columns):
    sns.boxplot(x='IncomeCategory', y=column, data=df, ax=axes[i])
    axes[i].set_title(f'{column} by Income Category')
    axes[i].set_xlabel('Income Category')
    axes[i].set_ylabel(column)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
