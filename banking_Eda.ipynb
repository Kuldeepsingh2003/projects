#Job Analysis: Which job types are most common among clients? How does job type relate to the likelihood of subscribing to a term deposit?

```python
df2.job.value_counts()
```

```python
df1.job.value_counts().plot(kind='bar')
plt.xlabel('job')
plt.ylabel('count')
plt.title('job distribution')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.countplot(y='job', data=df1, order=df1['job'].value_counts().index)
plt.title('Job Type Distribution')
plt.xlabel('Count')
plt.ylabel('Job')
plt.show()

# Subscription rates by job type
job_subscription = df1.groupby('job')['y'].value_counts(normalize=True).unstack()
plt.figure(figsize=(12, 6))
job_subscription.plot(kind='bar', stacked=True)
plt.title('Subscription Rates by Job Type')
plt.xlabel('Job')
plt.ylabel('Proportion')
plt.show()
```

```python
sns.catplot(x='job',hue='y',data=df1,kind='count')
plt.xticks(rotation=90)
plt.xlabel('job')
plt.ylabel('count')
plt.title('job distribution')
plt.show()
```

#Marital Status: How does marital status affect the likelihood of subscribing to a term deposit? Are there significant differences in subscription rates between different marital statuses?

```python
df1.marital.value_counts()
```

```python
df1 = df1[df1['marital'] != 'unknown']
sns.catplot(x='marital',hue='y',data=df1,kind='count')
plt.xticks(rotation=90)
plt.xlabel('matial status')
plt.ylabel('count')
plt.title('job distribution')
plt.show()
```

```python
df1 = df1[df1['marital'] != 'unknown']
df1.dropna(subset=['marital'])
plt.figure(figsize=(8, 6))
sns.countplot(x='marital', data=df1, order=df1['marital'].value_counts().index)
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()

# Subscription rates by marital status
marital_subscription = df1.groupby('marital')['y'].value_counts(normalize=True).unstack()
plt.figure(figsize=(8, 6))
marital_subscription.plot(kind='bar', stacked=True)
plt.title('Subscription Rates by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Proportion')
plt.show()
```
